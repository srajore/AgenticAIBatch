```python
#imports
from typing import TypedDict
from langgraph.graph import StateGraph,START,END
```

```python
class TempratureState(TypedDict):
  celcius: float
  fahrenheit: float
```

```python
def convert_temprature(state: TempratureState):
  state['fahrenheit'] = state['celcius'] * 9/5 + 32
  return state
```

```python
workflow =StateGraph(TempratureState)
```

```python
workflow.add_node("convert_temprature",convert_temprature)


```

```python
workflow.add_edge(START,"convert_temprature")
workflow.add_edge("convert_temprature",END)
```

```python
app =workflow.compile()
```

```python
result = app.invoke({'celcius':40})
```
